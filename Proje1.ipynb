```python
import pandas as pd
def not_hesaplama(satir):
    
    satir=satir[:-1]
    liste=satir.split(":")
    ogrenci_ismi=liste[0]
    notlar=liste[1].split(",")
    
    vize=int(notlar[1])
    final=int(notlar[2])
 
    ortalama= vize * (40/100) + final * (60/100)
    
    if ortalama>=90 and ortalama<=100:
        harf = "AA GEÇTİ"
    elif ortalama>=85 and ortalama<=89:
        harf = "BA GEÇTİ"
    elif ortalama>=80 and ortalama<=84:
        harf = "BB GEÇTİ"
    elif ortalama>=75 and ortalama<=79:
        harf = "CB GEÇTİ"
    elif ortalama>=70 and ortalama<=74:
        harf = "CC GEÇTİ"
    elif ortalama>=65 and ortalama<=69:
        harf = "DC KOŞULLU GEÇTİ"
    elif ortalama>=60 and ortalama<=64:
        harf = "DD KOŞULLU GEÇTİ"
    elif ortalama>=50 and ortalama<=59:
        harf = "FD KALDI"
    else:
        harf = "FF KALDI"
    
    return ogrenci_ismi+ " - Olasılık dersi harf notu " + ":" + harf +"\n"
    

def ortalamalari_oku():
    with open("ogrenci_notları.txt","r",encoding="utf-8") as file:
        for satir in file:
            print(not_hesaplama(satir))

def not_Girisi():   
    print("***ÖNEMLİ BİLGİ : Vize notu %40, final notu %60 etki etmektedir!***" "\n")

    isim=input("ÖĞRENCİNİN ADI: ")
    soyad=input("ÖĞRENCİNİN SOY ADI: ")
    ogrenci_no=input("ÖĞRENCİNİN NUMARASI ""\n")
    
    print("\n"f'SİSTEME HOŞGELDİNİZ! SAYIN {ogrenci_no} NUMARALI ÖĞRENCİMİZ {isim} {soyad}'"\n")
    print("\n""DERSİ GEÇİP GEÇMEDİĞİNİZİ GÖRMEK İÇİN LÜTFEN SINAV PUANLARINIZI GİRİNİZ!""\n")

    vize=input("ÖĞRENCİNİN VİZE SINAVI: ")
    final=input("ÖĞRENCİNİN FİNAL SINAVI: ")

    
    with open("ogrenci_notları.txt","a",encoding="utf-8") as file:
        file.write(isim+" "+soyad+":"+ogrenci_no+","+vize+","+final+"\n")
    
def not_Kaydet():
    with open("ogrenci_notları.txt","r",encoding="utf-8") as file:
        liste=[]
        for i in file:
            liste.append(not_hesaplama(i))
            
    with open ("harf_notları.txt","w",encoding="utf-8") as file2:
        for i in liste:
            file2.write(i)
            
def frame():
    with open("harf_notları.txt","r",encoding="utf-8") as file:
        veri=pd.DataFrame(file)
        print(veri)
    
def excel():
    with open("harf_notları.txt","r",encoding="utf-8") as file:
        veri=pd.DataFrame(file)
        veri.to_excel("excel_harfNotları.xlsx")
    
    
            
    
    
print(" ------------ ÖĞRENCİ NOT SİSTEMİ ------------ ")
while True:
    secim=input("\n""1- Öğrenci Bilgi Girişi\n2- Notları Gör\n3- Notları Kaydet\n4- Data Frame Şeklinde Göster\n5- Excele Dönüştür\n6- Çıkış Yap\n\n")
    
    if int(secim)==1:
        not_Girisi()

    elif int(secim)==2:
        ortalamalari_oku()

    elif int(secim)==3:
        not_Kaydet()
        print("NOTLAR KAYDEDİLDİ""\n")
        
    elif int(secim)==4:
        frame()
        
    elif int(secim)==5:
        excel()
        
        

    else:
        print("\n""SİSTEMDEN ÇIKIŞ YAPILDI.")
        break

```
